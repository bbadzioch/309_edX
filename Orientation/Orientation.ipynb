# Sign of the determinant

Let ${\bf v}_1$ and ${\bf v}_2$ be two vectors in $\mathbb{R}^2$ and let $M$ be the $2\times 2$ matrix with 
${\bf v}$ and ${\bf w}$ as columns:

$$M = \begin{bmatrix}{\bf v}_1 & {\bf v}_2 \end{bmatrix}$$

Recall that the sign of the determinant of $M$ can be interpreted as follows:

- If the vector ${\bf v}_2$ is in the counterclockwise direction from ${\bf v}_1$ then $\det M > 0$. 
- If the vector ${\bf v}_2$ is in the clockwise direction from ${\bf v}_1$ then $\det M < 0$. 

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

A = np.array([0, 0])
B = np.array([4, 2])
C = np.array([0, 3])
D = np.array([3, -1])
AB = B - A
AC = C - A
AD = D - A
plt.style.use("seaborn")

plt.figure(figsize = (10, 5))
####
ax1 = plt.subplot(1,2,1)
plt.xticks(range(-5, 8, 2), labels = [])
plt.yticks(range(-5, 8, 2), labels = [])
ax1.set_aspect(1)
plt.xlim(-2, 5)
plt.ylim(-1.5, 3.5)
plt.arrow(*A, *(AB) , 
          head_width=0.2, 
          head_length=0.2, 
          ec='tomato',
          fc='tomato',
          lw = 4, 
          length_includes_head=True, 
          zorder=10)
plt.arrow(*A, *(AC) , 
          head_width=0.2, 
          head_length=0.2, 
          ec='steelblue', 
          fc='steelblue', 
          lw = 4, 
          length_includes_head=True, 
          zorder=10)


plt.plot(*zip(A),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2, zorder=20)
style="Simple,tail_width=0.5,head_width=8,head_length=8"
kw = dict(arrowstyle=style, color="k", lw = 2)
a = patches.FancyArrowPatch( A + 1.5*AB/np.linalg.norm(AB), 
                              A + 1.5*AC/np.linalg.norm(AC), 
                             connectionstyle="arc3,rad=0.3", **kw)
ax1.add_patch(a)
plt.text(*[1.9, 0.45] , r"${\bf v}_1$", color = "tomato", ha= "center", size=14)
plt.text(*[-0.5, 1.2] , r"${\bf v}_2$", color = "steelblue", ha= "center", size=14)
plt.title(r"$\det\  [\ {\bf v}_1 \ \   {\bf v}_2\ ] > 0$" , fontsize = 14 , y = 1.05)


####
ax2 = plt.subplot(1,2,2)
plt.xticks(range(-5, 8, 2), labels = [])
plt.yticks(range(-5, 8, 2), labels = [])
ax2.set_aspect(1)
plt.xlim(-2, 5)
plt.ylim(-1.5, 3.5)
plt.arrow(*A, *(AB) , 
          head_width=0.2, 
          head_length=0.2, 
          ec='tomato', 
          fc='tomato',
          lw = 4, 
          length_includes_head=True, 
          zorder=10)
plt.arrow(*A, *(AD) , 
          head_width=0.2, 
          head_length=0.2, 
          ec='steelblue',
          fc='steelblue', 
          lw = 4, 
          length_includes_head=True, 
          zorder=10)


plt.plot(*zip(A),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2, zorder=20)
style="Simple,tail_width=0.5,head_width=8,head_length=8"
kw = dict(arrowstyle=style, color="k", lw = 2)


b = patches.FancyArrowPatch(A + 1.5*AB/np.linalg.norm(AB), 
                            A + 1.5*AD/np.linalg.norm(AD), 
                             connectionstyle="arc3,rad=-0.3", **kw)

ax2.add_patch(b)
plt.text(*[1.5, 1.1] , r"${\bf v}_1$", color = "tomato", ha= "center", size=14)
plt.text(*[1.2, -0.95] , r"${\bf v}_2$", color = "steelblue", ha= "center", size=14)
plt.title(r"$\det\  [\ {\bf v}_1 \ \   {\bf v}_2\ ] < 0$", fontsize = 14, y = 1.05)
plt.tight_layout()
plt.savefig("orientation1.svg", bbox_inches='tight')
plt.show()
```

## 1. Sides of a line

For any two points $A$, $B$ on a plane there is a unique straight line passing through these points. Such line divides the plane into two half-planes. If $X$, $Y$ are points which do not belong to the line, then we can ask if these points are on the same side of the line (i.e. they belong to the same half-plane), or it they are on different sides. This problem can be solved as follows. 

* Let ${\bf v }_{AB}$ denote the vector which starts at the point $A$ and ends at the point $B$. Similarly, denote by ${\bf v }_{AX}$  the vector which starts at $A$ and ends at $X$ and by ${\bf v }_{AY}$ the vector which starts 
at $A$ and ends at $Y$. 

* Let $M_X$, $M_Y$ be the $2\times 2$ matrices given by 
$$
M_X = \begin{bmatrix}{\bf v }_{AB} & {\bf v }_{AX} \end{bmatrix}
\hskip 10mm
M_Y = \begin{bmatrix}{\bf v }_{AB} & {\bf v }_{AY} \end{bmatrix}
$$
  
* If the sign of $\det M_X$ is the same as that of  $\det M_Y$ then $X$ and $Y$ are on the same side of the line. If these determinants have different signs then these points are on different sides. 

This procedure works, because all vectors starting at the point $A$ and ending on one side of the line are positioned in the clockwise direction from the vector ${\bf v }_{AB}$, while all vectors starting at $A$ and ending on the opposite side of the line are in the counterclockwise direction from ${\bf v }_{AB}$. 

```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
A = np.array([-1, -0])
B = np.array([4, 2])
X = np.array([0, 3])
Y = np.array([2, -1])
AB = B - A
AX = X - A
AY = Y - A
plt.figure(figsize = (7, 5))
plt.axes(aspect="equal")
plt.style.use("seaborn")
plt.yticks(range(-2, 6))
plt.xticks(range(-5, 8))
plt.xlim(-4.5, 7)
plt.ylim(-1.5, 4)
#plt.grid()
plt.plot(*zip((AB*(-2) + A), (AB*(3) + A)), "-", color="tomato", lw = 2, alpha=0.7, zorder = 1)


plt.plot(*zip(X),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2, zorder=20)
plt.plot(*zip(Y),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2, zorder=20)
plt.plot(*zip(A, B),  "o", markerfacecolor='w', markeredgecolor="tomato",  ms=10, markeredgewidth = 2, zorder=20)

plt.text(*(A + np.array((-0.5, 0.15))) , "A", ha= "center", size=14)
plt.text(*(B + np.array((0, 0.3))) , "B", ha= "center", size=14)
plt.text(*(X + np.array((0,  0.35))) , "X", ha= "center", size=14)
plt.text(*(Y + np.array((0.4,  0))) , "Y", ha= "left", va = "center", size=14)


plt.savefig("orientation7.svg", bbox_inches='tight')
plt.show()
```

### Example

In the picture above we have:

$$A = (-1, 0),\ \ \ \ B = (4, 2),\ \ \ \ X = (0, 3),\ \ \ \ Y = (2, -1)$$

This gives:

$$
{\bf v}_{AB}
=
\begin{bmatrix}
5 \\
2 \\
\end{bmatrix},
\hskip 5mm
{\bf v}_{AX}
=
\begin{bmatrix}
1 \\
3 \\
\end{bmatrix},
\hskip 5mm
{\bf v}_{AY}
=
\begin{bmatrix}
\phantom{-}3 \\
-1 \\
\end{bmatrix}
$$

Therefore 
$$
\det M_X
= 
\det
\begin{bmatrix}
5 & 1 \\
2 & 3 \\
\end{bmatrix}
= 13
\hskip 10mm
\det M_Y
= 
\det
\begin{bmatrix}
5 & \phantom{-}3 \\
2 & -1 \\
\end{bmatrix}
= -11
$$

Since these two determinants have different signs, points $X$ and $Y$ are on different sides of the line passing through the points $A$ and $B$.

## 2. Intersection of  line segments

Let $A$, $B$, $C$, $D$ be points on the plane. Let $AB$ denote the line segment joining the points 
$A$ and $B$ and $CD$ be the line segment joining the points $C$ and $D$. We can determine is
the line segments $AB$ and $CD$ intersect as follows:

* Check if the points  $A$ and $B$ are on different sides of the line passing through $C$ and $D$.
* Check if the points  $C$ and $D$ are on different sides of the line passing through $A$ and $B$.

We can verify if these conditions are satisfied or not using the procedure described in Part 1. 
If both of these conditions are satisfied then the line segments intersect. 

```python
import matplotlib.pyplot as plt
import numpy as np
A = np.array([-1, 2])
B = np.array([4, -1])
C = np.array([-2, -1])
D = np.array([4, 2])
AB = B-A
CD = D-C
plt.style.use("default")
plt.plot(*zip(A, B), color = "tomato", lw = 5)
plt.plot(*zip(C, D) , color = "steelblue", lw = 5)
plt.xticks([])
plt.yticks([])
plt.xlim(-4.5, 7)
plt.ylim(-3, 4)
plt.plot(*zip((AB*(-2) + A), (AB*(3) + A)), "k-", lw = 2, alpha=0.4, zorder = 1)
plt.plot(*zip((CD*(-2) + C), (CD*(3) + C)), "k-", lw = 2, alpha=0.4, zorder = 1)

plt.plot(*zip(A, B),  "o", markerfacecolor='w', markeredgecolor="tomato",  ms=10, markeredgewidth = 2)
plt.plot(*zip(C, D),  "o", markerfacecolor='w', markeredgecolor="steelblue",  ms=10, markeredgewidth = 3)
plt.text(*(A + np.array((0, 0.5))) , "A", ha= "center", size=14)
plt.text(*(B + np.array((0.05, -0.4))) , "B", ha= "center", va = "top", size=14)
plt.text(*(D + np.array((0, 0.5))) , "D", ha= "center", size=14)
plt.text(*(C + np.array((-0.05, -0.5))) , "C", ha= "center", va = "top", size=14)
plt.savefig("orientation3.svg", bbox_inches='tight')
plt.show()
```

## 3. Detecting points inside a triangle. 

Let $A$, $B$, $C$ be points on the plane forming a triangle. Given a point $X$, we can determine if it is inside or outside of this triangle as follows. 

* Check if the points $X$ and  $A$ on the same side of the line passing through $B$ and $C$.
* Check if the points $X$ and  $B$ on the same side of the line passing through $A$ and $C$.
* Check if the points $X$ and  $C$ on the same side of the line passing through $A$ and $B$.

Again, we can use here the procedure described in Part 1 to check if points are on the same side of a line or not.
If all of these conditions are satisfied then the point $X$ is inside of the triangle 
$ABC$, otherwise it is outside

```python
import matplotlib.pyplot as plt
import numpy as np
A = np.array([0.7, 2])
B = np.array([-1, -1])
C = np.array([3, 0])
X = np.array([0.7, 0.5])
AB = B-A
AC = C-A
BC = C-B
plt.style.use("default")
plt.plot(*zip((AB*(-2) + A), (AB*(3) + A)), "k-", lw = 2, alpha=0.4, zorder = 1)
plt.plot(*zip((AC*(-2) + A), (AC*(3) + A)), "k-", lw = 2, alpha=0.4, zorder = 1)
plt.plot(*zip((BC*(-2) + B), (BC*(3) + B)), "k-", lw = 2, alpha=0.4, zorder = 1)
plt.fill(*zip(A, B, C),  alpha=0.3, color = "tomato")
plt.fill(*zip(A, B, C), fill=None, color = "tomato", lw = 5, zorder = 10)
plt.plot(*zip(A, B, C),  
         "o", 
         markerfacecolor='w', 
         markeredgecolor="tomato",  
         ms=10, 
         markeredgewidth = 2, 
         zorder =20)

plt.xticks([])
plt.yticks([])
plt.xlim(-3, 5.5)
plt.ylim(-2, 4)
plt.plot(*zip(X),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2)
plt.text(*(A + np.array((0, 0.35))) , "A", ha= "center", size=14)
plt.text(*(B + np.array((0.05, -0.3))) , "B", ha= "center", va = "top", size=14)
plt.text(*(X + np.array((0, 0.35))) , "X", ha= "center", size=14)
plt.text(*(C + np.array((-0.05, -0.3))) , "C", ha= "center", va = "top", size=14)
plt.savefig("orientation4.svg", bbox_inches='tight')
plt.show()
```

## 4. Detecting points inside a polygon


Let $P$ be a polygon on the plane with some number of vertices, and let $X$ be point on the plane. 
We can determine if $X$ is inside or outside of the polygon as follows. 

* Select a point $Y$ which is outside of the polygon (for example, you can pick a point whose $x$-cordinate is larger than $x$-coordinates of all vertices of the polygon)
* Check how many sides of the polygon are intersected by the line segment $XY$ (we can do it using the method from  Part 2). 

If the number of sides intersecting with $XY$ is odd, then the point $X$ is inside the polygon. Otherwise $X$ is outside.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize= (5.5, 4))

A = np.array([0, 0])
B = np.array([2, 3])
C = np.array([3, 0])
D = np.array([4, 4])
E = np.array([-1, 4])
X = np.array([0.3, 2])
Y = np.array([5, 2])
C_1 = np.array([4/3, 2])
C_2 = np.array([2+(1/3), 2])
C_3 = np.array([3+(1/2), 2])
AB = B-A
AC = C-A
BC = C-B
plt.style.use("default")
plt.fill(*zip(A, B, C, D, E),  alpha=0.3, color = "tomato")
plt.fill(*zip(A, B, C, D, E), fill=None, color = "tomato", lw = 5, zorder = 10)
plt.plot(*zip(A, B, C, D, E),  
         "o", 
         markerfacecolor='w', 
         markeredgecolor="tomato",  
         ms=10, 
         markeredgewidth = 2, 
         zorder =20)

plt.xticks([])
plt.yticks([])
plt.xlim(-2.5, 6)
plt.ylim(-1, 5.5)
plt.plot(*zip(X, Y), "k-", lw = 4, alpha = 0.5, zorder = 10)
plt.plot(*zip(X, Y),  "o", markerfacecolor='w', markeredgecolor="k",  ms=10, markeredgewidth = 2, zorder=20)
plt.plot(*zip(C_1),  "o", markerfacecolor='k', markeredgecolor="k",  ms=10, markeredgewidth = 0, zorder=20)
plt.plot(*zip(C_2),  "o", markerfacecolor='k', markeredgecolor="k",  ms=10, markeredgewidth = 0, zorder=20)
plt.plot(*zip(C_3),  "o", markerfacecolor='k', markeredgecolor="k",  ms=10, markeredgewidth = 0, zorder=20)



plt.text(*(X + np.array((0, 0.35))) , "X", ha= "center", size=14)
plt.text(*(Y + np.array((0, 0.35))) , "Y", ha= "center", size=14)
plt.savefig("orientation5.svg", bbox_inches='tight')
plt.show()
```

**Note:** In order to complete this procedure we also need to specify how to handle two special cases:

1. It may happen that the line segment $XY$ will pass though a vertex of the polygon. 
2. It may also happen a part of the segment $XY$ will overlap with one of the edges of the polygon. 


In the first case, we look at the two vertices of the polygon (marked blue in the picture) neighboring the intersecting vertex. If these vertices are on the same side of the line $XY$, we count the intersection point as a double intersection. If they are on the different sides, we count it as a single intersection. 

The second case is similar. We look at the two vertices of the polygon (marked blue in the picture) that are neighboring the intersecting edge. Again, we count the intersection as either a double or a single intersection depending if these vertices are on the same side or different sides of the line $XY$.
