```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib notebook
```

```python
img = plt.imread("green_screen.jpg")
plt.figure(figsize = (8, 8))
plt.imshow(img)
plt.show()
```

```python
imgc = img.copy()
img2 = abs(img  - np.array([127, 182,  89])).sum(axis=2)
img3 = abs(img  - np.array([ 82, 136,  50])).sum(axis=2)
img4 = abs(img  - np.array([38, 64, 17])).sum(axis=2)
imgc[(img2 <80) | (img3 <80) |  (img4 < 40)] = 255
plt.figure(figsize = (8, 8))
plt.imshow(imgc)
plt.show()
```

```python
import cv2
def show_img(img):
    plt.figure(figsize = (8, 8))
    imgc = img[:, :, [2, 1, 0]]
    plt.imshow(imgc)
    plt.show()
```

```python
imgc = img.copy()
img2 = img[:, :, 1]/(img.sum(axis=2) + 0.01)
imgc[img2 > 0.38] = [0, 0, 100]
show_img(imgc)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("some_words.csv")
df
```

```python
df["pointless"] = 0
df
```

```python
df["length"] = df["phrase"].map(len)
```

```python
df["first capital"] = df["phrase"].map(lambda s : s[0] == s[0].upper())
df
```

```python
df["vowels"] = df["phrase"].map(lambda s : len([x for x in s if x  in "aeiou"]))
df
```

```python
df["vowels_ratio"] = df["vowels"]/df["length"]
df
```

```python
df["reversal"] = df["phrase"].map(lambda s : s[::-1])
df
```

```python
df["is_palindrome"] = (df["phrase"] == df["reversal"])
df
```

```python
df["sorted"] = df["phrase"].map(lambda s: "".join(sorted(s)))
df
```

```python
df["has_reversal"] = df["phrase"].map(lambda s: s in df["reversal"].values)
df
```
