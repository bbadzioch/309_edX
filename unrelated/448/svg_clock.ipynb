```python
from time import sleep
import datetime
from IPython.display import SVG, clear_output
s = '''<svg xmlns="http://www.w3.org/2000/svg" width="1000" height="150">
    <text x="50" y="120" stroke="#606060" font-size="140" stroke-width="3" fill="#bbaaff">{}</text>
</svg>'''
while True:
    currentDT = datetime.datetime.now()
    t = currentDT.strftime("%H:%M:%S")
    display( SVG(s.format(t)))
    sleep(1)
    clear_output(wait=True)
```
